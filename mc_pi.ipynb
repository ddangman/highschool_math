Equation for circle: x^2 + y^2 = r^2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample 10,000 random numbers from a uniform distribution between -1 and 1
df = pd.DataFrame(np.random.uniform(-1, 1, size=(10000, 2)), columns=list('XY'))
```

```python
plt.scatter(df['X'], df['Y'])
plt.axis('square')
```

```python
# function that checks if a point is inside the circle
def is_inside_circle(x, y):
    return x**2 + y**2 <= 1
```

```python
in_circle = list(map(is_inside_circle, df['X'], df['Y']))
df.insert(2, 'in_circle', in_circle)
```

```python
plt.scatter(df['X'], df['Y'], c=df['in_circle'])
plt.axis('square')
```

```python
df['in_circle'].sum()
```

```python
# function that checks if point is in first quadrant
def is_in_first_quadrant(x, y):
    return x >= 0 and y >= 0
```

```python
first_quadrant = list(map(is_in_first_quadrant, df['X'], df['Y']))
df.insert(2, 'in_square', first_quadrant)
```

```python
plt.clf()
plt.scatter(df['X'], df['Y'], c=df['in_square'])
plt.axis('square')
```

```python
df['in_circle'].sum() / df['in_square'].sum()
```
